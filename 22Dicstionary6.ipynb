# Defaultdict
**when value is not given it assumes as 0**

```python
from collections import defaultdict

name=['Virat kholi','Dinesh Kartik','AB Devillers','Dhoni ']
grouped= defaultdict(list)

for n in name:
  grouped[n[0]].append(n)

print(dict(grouped))
```
